```python
world_vacc = df["vaccines"].str.split(",",expand=True)
world_vacc.columns = ["vac_1","vac_2","vac_3","vac_4","vac_5","vac_6","vac_7","vac_8",]
world_vacc
```

```python
world_vacc.isnull().sum()
```

```python
world_vacc_v1 = pd.concat([world_vacc["vac_1"], df["country"]],axis=1)
col_name = ["Vaccine_name", "Country"]
world_vacc_v1.columns = col_name

world_vacc_v1
```

```python

for i, column in enumerate(world_vacc.columns):
    if i!=0:
        col = world_vacc.columns[i]
        vacc_temp = pd.concat([world_vacc[world_vacc.columns[i]], df["country"]], axis=1)
        vacc_temp.columns = col_name
        
        world_vacc_v1 = pd.concat([world_vacc_v1, vacc_temp], ignore_index=True)
        del vacc_temp
        
world_vacc_v1
```

```python
world_vacc_v1["Vaccine_name"] = world_vacc_v1["Vaccine_name"].str.strip()
world_vacc_v1
```
