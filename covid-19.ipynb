```python
vaccine_source = {}

for source in df.source_name.unique():
    count = df.groupby(df["source_name"]==source)["country"].nunique()
    vaccine_source[source] = count[True]
    
vaccine_source_df = pd.DataFrame.from_dict(vaccine_source, orient='index')
vaccine_source_df.columns = ["Count"]
vaccine_source_df.sort_values(by="Count", ascending=False, inplace=True)
vaccine_source_df.head(20)
```

```python
vaccine_source_df.iloc[:10].plot(kind="bar")
plt.title("Source of Vaccine Suppliers")
plt.ylabel("Number of vaccine provided in country")
```

From the above chart we can conclude that World Health Organization is the highest vaccine source to multiple countries
followed by Ministry of Health and SPC Public Health Division

Daily Vaccination till date in each country

```python
vaccinated_in_country = {}
for country in df.country.unique():
    vaccinated_count = 0
    for i in range(len(df[df["country"]==country])):
        vaccinated_count += df["daily_vaccinations"][i]
        
    vaccinated_in_country[country] = vaccinated_count

vaccinated_in_country_df = pd.DataFrame.from_dict(vaccinated_in_country, orient="index")
vaccinated_in_country_df.columns = ["vaccinated_till_date"]
vaccinated_in_country_df.sort_values(by="vaccinated_till_date", ascending=False, inplace=True)
vaccinated_in_country_df.head(20)
```

```python
vaccinated_in_country_df.iloc[:10].plot(kind="bar")
plt.title("Top 10 most vaccinated country in world")
plt.ylabel("Number of citizens vaccinated (per million)")
```
